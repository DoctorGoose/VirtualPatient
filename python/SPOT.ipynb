```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '../Misc/DAD04000.xlsx'
xls = pd.ExcelFile(file_path)
sheet_names = xls.sheet_names

def process_and_plot_all_patients(xls, sheet_names):
    param_columns = ['', 'a', '', 'E', 'KE']
    metric_columns = ['DOF', 'AIC']
    
    for sheet_name in sheet_names:
        # Load the sheet
        data = pd.read_excel(xls, sheet_name=sheet_name)
        
        # Extract the parameter and metric names from the first row and assign them as the column headers
        new_column_names = data.iloc[0].tolist()
        data.columns = new_column_names
        data = data.drop(0)
        
        # Convert to numeric, forcing errors to NaN (so we can skip NaN values later)
        data[param_columns + metric_columns] = data[param_columns + metric_columns].apply(pd.to_numeric, errors='coerce')
        
        # Prepare the figure and subplots
        fig, axs = plt.subplots(len(param_columns), 1, figsize=(10, 20), sharex=True)
        fig.suptitle(f'Patient {sheet_name} - AIC vs DOF by Parameter')
        
        for i, param in enumerate(param_columns):
            # Filter data based on the parameter being fit (non-zero)
            individual_data = data[data[param].notna() & (data[param] != 0)]
            mp_data = data[data[param].isna() | (data[param] == 0)]
            
            # Plot for individual fits
            axs[i].scatter(individual_data['DOF'], individual_data['AIC'], color='blue', label=f'{param} - Individual')
            # Plot for MP (zero fits)
            axs[i].scatter(mp_data['DOF'], mp_data['AIC'], color='orange', label=f'{param} - MP')
            
            axs[i].set_ylabel('AIC')
            axs[i].legend()
            axs[i].set_title(param)
        
        axs[-1].set_xlabel('Degrees of Freedom (DOF)')
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the title
        plt.show()

# Process and plot for all patients
process_and_plot_all_patients(xls, sheet_names)
```

```python
import statsmodels.api as sm
cleaned_data = data.dropna(subset=['AIC', 'SSE'])

# Fit the OLS model using statsmodels to calculate confidence intervals
X = (cleaned_data[['', 'T0', 'p', 'c']] != 0).astype(int)
X = sm.add_constant(X)  # Adds the intercept term
y_aic = cleaned_data['AIC']
y_sse = cleaned_data['SSE']

# Fit the models
model_aic = sm.OLS(y_aic, X).fit()
model_sse = sm.OLS(y_sse, X).fit()

# Get the confidence intervals
ci_aic = model_aic.conf_int(alpha=0.05)
ci_sse = model_sse.conf_int(alpha=0.05)

# Prepare the DataFrame for visualization
coefficients_with_ci = pd.DataFrame({
    'Parameter': ['Intercept', '', 'T0', 'p', 'c'],
    'AIC Coefficient': model_aic.params,
    'AIC CI Lower': ci_aic[0],
    'AIC CI Upper': ci_aic[1],
    'SSE Coefficient': model_sse.params,
    'SSE CI Lower': ci_sse[0],
    'SSE CI Upper': ci_sse[1]
})

print(coefficients_with_ci)
plt.figure(figsize=(12, 6))
plt.errorbar(coefficients_with_ci['Parameter'], coefficients_with_ci['AIC Coefficient'],
             yerr=[coefficients_with_ci['AIC Coefficient'] - coefficients_with_ci['AIC CI Lower'], 
                   coefficients_with_ci['AIC CI Upper'] - coefficients_with_ci['AIC Coefficient']],
             fmt='o', color='blue', ecolor='lightblue', elinewidth=3, capsize=5, label='AIC Coefficient')

plt.errorbar(coefficients_with_ci['Parameter'], coefficients_with_ci['SSE Coefficient'],
             yerr=[coefficients_with_ci['SSE Coefficient'] - coefficients_with_ci['SSE CI Lower'], 
                   coefficients_with_ci['SSE CI Upper'] - coefficients_with_ci['SSE Coefficient']],
             fmt='o', color='red', ecolor='lightcoral', elinewidth=3, capsize=5, label='SSE Coefficient')

# Labels and title
plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel('Parameter')
plt.ylabel('Coefficient Value')
plt.title('Regression Coefficients with 95% CI')
plt.legend()
plt.grid(True)
plt.show()
```
